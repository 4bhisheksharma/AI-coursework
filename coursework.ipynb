```python
# import pandas & parquet
import pandas as pd
import numpy as np
from pandas.io import parquet
```

```python
# file path for drive
parquet_file_path  = 'drive/MyDrive/Colab Notebooks/movies.parquet'

# file path for local
# parquet_file_path  = 'movies.parquet'
```

```python
#import parquet
df = pd.read_parquet(parquet_file_path)
df.head()
```

```python
#to get some info about dataset
df.info()
```

```python
# to find the shape
print("Dataset Shape:", df.shape)
```

```python
# checking missing values
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100
missing_df = pd.DataFrame({
    'Column': missing.index,
    'Missing Count': missing.values,
    'Percentage': missing_percent. values
})
print(missing_df[missing_df['Missing Count'] > 0]. sort_values('Missing Count', ascending=False))
```

```python
# Cleaning the dataset
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python
# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cell 6: Visualize rating distribution

plt.figure(figsize=(15, 5))

# Vote Average Distribution
plt.subplot(1, 3, 1)
plt.hist(df['vote_average']. dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Movie Ratings')
plt.xlabel('Vote Average')
plt.ylabel('Frequency')

# Vote Count Distribution (log scale)
plt.subplot(1, 3, 2)
plt.hist(np.log1p(df['vote_count']. dropna()), bins=30, color='salmon', edgecolor='black')
plt.title('Distribution of Vote Counts (Log Scale)')
plt.xlabel('Log(Vote Count)')
plt.ylabel('Frequency')

# Popularity Distribution (log scale)
plt.subplot(1, 3, 3)
plt.hist(np.log1p(df['popularity'].dropna()), bins=30, color='lightgreen', edgecolor='black')
plt.title('Distribution of Popularity (Log Scale)')
plt.xlabel('Log(Popularity)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
